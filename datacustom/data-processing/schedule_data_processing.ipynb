```python
# label을 확인하여 알맞은 라벨을 넣어줌
def getLabel(b_label):
    label = 'B-'
    label_list = []

    while len(b_label):
        labelling = b_label.pop()
        if len(b_label) == 0:
            if label == 'B-':
                label = 'S-'
            else:
                label = 'E-'

        labelling = label+labelling
        label_list.append(labelling)
        if label == 'B-':
            label = 'I-'
            
    return label_list
```
